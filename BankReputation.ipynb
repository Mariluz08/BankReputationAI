```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()
# Leer el archivo Excel
data = pd.read_excel('Menciones_Octubre_9_12_Bancolombia_2023.xlsx')

# Asegurarse de que la columna 'Fecha' esté en formato de fecha
data['Fecha'] = pd.to_datetime(data['Fecha'])

# Crea la columna 'Hora' a partir de la columna 'Fecha'
data['Hora'] = data['Fecha'].dt.hour

# Mapea las categorías de sentimiento a valores numéricos
sentimiento_mapping = {'NEGATIVE': -1, 'NEUTRAL': 0, 'POSITIVE': 1}
data['Sentimiento'] = data['Sentimiento'].map(sentimiento_mapping)

# Verifica si la columna 'Hora' existe
if 'Hora' in data:
    hourly_sentiment = data.groupby('Hora')['Sentimiento'].mean()

    # Crea un gráfico de barras de sentimiento promedio por hora
    plt.figure(figsize=(10, 6))
    hourly_sentiment.plot(kind='bar')
    plt.title('Sentimiento promedio por hora')
    plt.xlabel('Hora')
    plt.ylabel('Sentimiento promedio')
    plt.xticks(rotation=0)
    plt.show()
else:
    print("La columna 'Hora' no se ha creado correctamente en los datos.")
```
